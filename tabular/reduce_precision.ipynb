# Reduce size of hospital coordinates

This might save some file space when the data is repeated lots.

```python
import pandas as pd
```

```python
df_hospitals = pd.read_csv("../data_tabular/stroke_hospitals_2022.csv")

df_hospitals.head()
```

```python
df_hospitals['long'] = df_hospitals['long'].round(4)
df_hospitals['lat'] = df_hospitals['lat'].round(4)
```

```python
df_hospitals.to_csv('stroke_hospitals_2022_reduced.csv', index=False)
```
