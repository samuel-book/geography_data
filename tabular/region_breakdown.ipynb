```python
import numpy as np
import pandas as pd
```

```python
df_regions = pd.read_csv('../data_tabular/LSOA_regions.csv')
```

```python
region_cols = ['RGN11NM', 'SCN17NM', 'LAD17NM', 'LHB20NM', 'STP19NM', 'CCG19NM']

region_set = df_regions['RGN11NM'].unique()
for region in region_set:
    df_this_region = df_regions[df_regions['RGN11NM'] == region]
    if region != 'Wales':
        print(region)
        print(df_this_region['SCN17NM'].unique())
        print(df_this_region['STP19NM'].unique())
        print(df_this_region['LAD17NM'].unique())
        print(df_this_region['CCG19NM'].unique())
        print('\n')
    else:
        print(region)
        print(df_this_region['LAD17NM'].unique())
        print(df_this_region['LHB20NM'].unique())
        print('\n')
```
